```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob

figures = glob.glob('results/figures/*.png')
if not figures:
    print("No figures found. Make sure to run the interpretability step first!")
for fig in figures:
    plt.figure(figsize=(10, 5))
    img = mpimg.imread(fig)
    plt.imshow(img)
    plt.axis('off')
    plt.title(fig.split('/')[-1])
    plt.show()
```
